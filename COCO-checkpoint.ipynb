```python
def diff_choc(x):
    if x == 7:
        return 'BSSCVIS'
    elif x == 6:
        return 'BSSCVI'
    elif x == 5:
        return 'BSSCI'
    elif x == 4:
        return 'BSSC'
    elif x == 3:
        return 'BSS'
    elif x == 2:
        return 'B-SW'
    else:
        return 'SW'
```

```python
coco_cdf['n'] = coco_cdf['number_of_ingredients'].apply(diff_choc)
```

```python
cat = ['company_location','country_of_bean_origin','sweet','nutty','cocoa','roasty','earthy','creamy','sandy','fatty','floral','intense','spicy','sour','vanilla','fruit','molasses','woody','sticky','coffee','rich','dried.fruit','ingredient_Beans','ingredient_Sugar','ingredient_Sweetener','ingredient_Cocoa_Butter','ingredient_vanilla','ingredient_lecithin','ingredient_salt','n']
```

```python
from scipy.stats import zscore,chi2_contingency,chi2
cat_dep=[]
cat_indep = []
for name in cat:
    crt=pd.crosstab(coco_cdf['rating'],coco_cdf[name])
    stat, p, dof, expected = chi2_contingency(crt)
    prob = 0.95
    critical = chi2.ppf(prob, dof)
    if abs(stat) >= critical:
        cat_dep.append('{} : Dependent'.format(name))
    else:
        cat_dep.append('{} : Independent'.format(name))
        cat_indep.append(name)
cat_dep
```
